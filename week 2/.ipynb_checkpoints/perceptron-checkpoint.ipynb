# 3. The perceptron

```python
import numpy as np
X = np.array([[1, 2, 2], [1, 2, 3], [1, 4, 2], [1, 4, 0]])
Y = np.array([[1], [1], [-1], [-1]])
w = np.array([[0], [0], [0]])
i = 0
while True:
    x = X[[i], :]
    y = Y[i]
    print("---------------------")
    print(f"{i=}")
    print(f"{x=}")
    print(f"{y=}")
    print(f"w={w.flatten()}")
    print(f"calculation = {np.abs(np.sign(x @ w) - y)}")
    if np.abs(np.sign(x @ w) - y) > 1e-9:
        print("Misclassification")
        w += y * x.transpose()
        i = 0
    elif i == X.shape[0] - 1:
        print("Correct classification. Job done.")
        break
    else: 
        print("Correct classification")  
        i += 1
```
